```python
import pandas as pd

#Function to preprocess the dataset
def preprocess_data(file_path):
    data = []
    with open('/train_data.txt', 'r') as file:
        for line in file:
            parts = line.strip().split(' ::: ')
            if len(parts) == 4:  # Ensure the line has all required parts
                data.append({
                    'id': parts[0],
                    'title': parts[1],
                    'genre': parts[2],
                    'plot': parts[3]
                })
    return pd.DataFrame(data)

#Load the training data
train_data = preprocess_data('/train_data.txt')

#View the first few rows of the data
train_data.head()
```

```python
# Features: movie plots, Labels: genres
X = train_data['plot']
y = train_data['genre']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert text data into TF-IDF features
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_tfidf = tfidf_vectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Train a Naive Bayes model
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

# Make predictions on the test data
y_pred = model.predict(X_test)

# Print the classification report to evaluate the model
print(classification_report(y_test, y_pred))
```

```python
# Example: Predict genre for a new movie plot
new_plot = ["A young boy discovers he has magical powers and must attend a wizarding school."]
new_plot_tfidf = tfidf_vectorizer.transform(new_plot)
predicted_genre = model.predict(new_plot_tfidf)
print(predicted_genre)
```

```python
from sklearn.linear_model import LogisticRegression

# Train Logistic Regression model
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)

# Evaluate the model
y_pred_lr = lr_model.predict(X_test)
print(classification_report(y_test, y_pred_lr))
```

```python
# Preprocess the test data
test_data = preprocess_data('/test_data.txt')

# Separate features (plots) and labels (genres)
X_test_plots = test_data['plot']
y_test_actual = test_data['genre']
```

```python
# Transform the test plots into TF-IDF features
X_test_tfidf = tfidf_vectorizer.transform(X_test_plots)
```

```python
# Use the trained model to predict the genres
y_test_pred = model.predict(X_test_tfidf)
```

```python
# Evaluate the model's performance
print(classification_report(y_test_actual, y_test_pred))
```
